### Random Tensors

```python
# Create a random tensor of size (5, 3)
random_tensor = torch.randn(5, 3)
print(random_tensor)
```

```python
random_tensor.ndim
```

```python
# Create a random tensor with similar shape to an image tensor
random_image_size_tensor = torch.rand(size = (3, 224, 224))
random_image_size_tensor.shape, random_image_size_tensor.ndim
```

### Zeros and ones

```python
zeros = torch.zeros(size=(3, 4))
zeros
```

```python
zeros_like_random_tensor = torch.zeros_like(random_tensor)
ones_like_random_tensor = torch.ones_like(random_tensor)
```
